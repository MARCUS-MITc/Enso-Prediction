```python
import netCDF4 as nc
from netCDF4 import num2date,date2index
import datetime
import csv
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('ensod.csv')
df
```

```python
for column in df.columns:
    if column != 'year':
        month_mean = df[column].mean()
        month_std = df[column].std()
        
        # Normalize other months
        df[column] = (df[column] - month_mean) / month_std
```

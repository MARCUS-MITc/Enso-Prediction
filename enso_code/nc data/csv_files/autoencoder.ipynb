```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras.layers import Input, Dense
from keras.models import Model
from sklearn.preprocessing import normalize
```

```python
df = pd.read_csv('air_T.csv')
df
```

```python
for index, row in df.iterrows():
    month_data = row.iloc[1:]  # Exclude the first column (year)
    print(month_data)
```

```python
first_row_data = df.iloc[0,1:].values
first_row_data
```

```python
first_row_data.shape
```

```python
normalized_input_data = normalize(first_row_data.reshape(1, -1))
normalized_input_data
```

```python
normalized_input_data.shape
```

```python
np.max(np.abs(first_row_data))
```

```python
input_dim = 324
hidden_dim = 65
input_layer = Input(shape = (input_dim,))
hidden_layer = Dense(hidden_dim,activation='tanh')(input_layer)
output_layer = Dense(input_dim,activation='linear')(hidden_layer)
```

```python
autoencoder = Model(inputs=input_layer, outputs=output_layer)
```

```python
autoencoder.compile(optimizer='adam', loss='mean_squared_error')
```

```python
autoencoder.fit(normalized_input_data, normalized_input_data, epochs=10, batch_size=32)
```

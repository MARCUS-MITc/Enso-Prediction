```python
import netCDF4 as nc
from netCDF4 import num2date,date2index
import datetime
import csv
import numpy as np
```

```python
data = nc.Dataset('sst.nc')
```

```python
time = data.variables['time']
time_ = time[:]
time_units = time.units
time_calendar = time.calendar
time_dates = num2date(time_, units=time_units, calendar=time_calendar)
start_date = datetime.datetime(1958, 1, 1)
start_index = date2index(start_date, time, select='nearest')
slp_subset = data['olr'][start_index:, :, :]
```

```python
data_ = np.array(slp_subset)
data_[data_ == -9.96921e+36] = np.nan
```

```python
import numpy as np

# Assuming you have a numpy array named 'data' of shape (504, 73, 144)

# Reshape the data to have time and month dimensions separately
data_reshaped = np.reshape(data_, (len(data_) // 12, 12, 73, 144))

# Calculate the mean and standard deviation for each month of all years
mean_all_months = np.nanmean(data_reshaped, axis=(0, 2, 3))
std_all_months = np.nanstd(data_reshaped, axis=(0, 2, 3))

# Normalize the data for each month
normalized_data = np.empty_like(data_)
for month in range(12):
    month_data = data_reshaped[:, month]
    month_mean = mean_all_months[month]
    month_std = std_all_months[month]
    normalized_month_data = (month_data - month_mean)/month_std
    normalized_data = np.reshape(normalized_month_data, (len(data_) // 12, 73, 144))
    #normalized_data[:, month] = np.repeat(normalized_month_data,12,axis =0)

# Print the shape of the normalized data
print(normalized_data.shape)
```

```python
lat_avg_size = data_.shape[2] // 18
lon_avg_size = data_.shape[3] // 18
#level_avg_size = 1
# Calculate the number of latitude and longitude divisions for averaging
lat_divisions = 18
lon_divisions = 18
#level_divisions = data.shape[1] // level_avg_size
# Adjust longitude divisions to get 18 divisions
# if lon_divisions > 18:
#     lon_avg_size += 1
#     lon_divisions = longitude.size // lon_avg_size
# if lon_divisions > lat_divisions:
#     lon_divisions = lat_divisions

# Create an empty array to store the averaged data
averaged_data = np.zeros((lat_divisions, lon_divisions))

# Iterate over latitude and longitude ranges
for lat_idx in range(lat_divisions):
    for lon_idx in range(lon_divisions):
        lat_range_start = lat_idx * lat_avg_size
        lat_range_end = (lat_idx + 1) * lat_avg_size
        lon_range_start = lon_idx * lon_avg_size
        lon_range_end = (lon_idx + 1) * lon_avg_size

        # Extract the subset of data
        subset = normalized_data[:,0, lat_range_start:lat_range_end, lon_range_start:lon_range_end]

        # Calculate the average along the latitude, longitude, and time dimensions
        averaged_subset = np.mean(subset, axis=(0, 1, 2))

        # Store the averaged data
        averaged_data[lat_idx, lon_idx] = averaged_subset
```

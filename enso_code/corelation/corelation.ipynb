```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, kendalltau, spearmanr
from sklearn.feature_selection import mutual_info_regression
```

```python
predictors = pd.read_csv('single_predictors.csv')
predictors
```

```python
predictors.fillna(0)
```

```python
flat = predictors['airT'].values.flatten()
flat = flat[:21]
flat
```

```python
indices = pd.read_csv('enso_index.csv')
indices
```

```python
indices_flat = indices.values[:, 1]
indices_flat
```

```python
lead_months = range(1,13)
best_predictors = []
```

```python
for lead_month in lead_months:
    correlations = []
    for predictor in predictors:
        # Calculate correlation between predictor and indices for the lead month
        index_month = lead_month - 1
        #predictor_month = index_month - 1
        
        # Flatten the indices and predictor arrays for correlation calculation
        indices_flat = indices.iloc[:, index_month].values.flatten()
        
        # Calculate correlation measures
        pearson_corr = pearsonr(indices_flat, flat)[0] 
        kendall_corr = kendalltau(indices_flat, flat)[0]
        spearman_corr = spearmanr(indices_flat, flat)[0]
        mi = mutual_info_regression(indices_flat.reshape(-1, 1), flat.reshape(-1, 1))[0]
        
        # Store the correlation measures
        correlations.append((pearson_corr, kendall_corr, spearman_corr, mi))
        
         # Select the predictor with the highest correlation for each measure
    max_corr_index = np.argmax(correlations, axis=0)
    best_predictor = predictors[max_corr_index]
        
        # Store the best predictor for the lead month
    best_predictors.append(best_predictor)

print("Best predictors for each lead month:")
for lead_month, predictor in zip(lead_months, best_predictors):
    print(f"Lead Month: {lead_month}, Best Predictor: {predictor}")
```

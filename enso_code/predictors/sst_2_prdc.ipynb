```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.layers import Input, Dense
from keras.models import Model
from sklearn.preprocessing import normalize
```

```python
df = pd.read_csv("project 1.csv")
df
```

```python
slp = df['slp'].values
slp
```

```python
slpi = normalize(slp.reshape(1,-1))
slpi
```

```python
max_slpi_value = np.amax(np.absolute(slpi))
max_slpi_value
```

```python
slpi.max()
```

```python
sst = df['sst'].values
sst = sst[~np.isnan(sst)]
sst
```

```python
ssti = normalize(sst.reshape(1,-1))
ssti
```

```python
max_ssti_value = np.amax(np.absolute(ssti))
max_ssti_value
```

```python
ssti.max()
```

```python
slpi_flat = np.ravel(slpi)
ssti_flat = np.ravel(ssti)
```

```python
slpi_flat.shape
```

```python
ssti_flat.shape
```

```python
merged_list = []
#ssti_flat + slpi_flat
merged_list.extend(ssti_flat)
merged_list.extend(slpi_flat)
```

```python
merged_list
```

```python
input_data = np.array(merged_list)
input_data
```

```python
input_data.shape
```

```python
input_data = input_data.reshape((1, 456))
```

```python
input_dim = 456
hidden_dim = 45
input_layer = Input(shape = (input_dim,))
hidden_layer = Dense(hidden_dim,activation='tanh')(input_layer)
output_layer = Dense(input_dim,activation='linear')(hidden_layer)
```

```python
autoencoder = Model(inputs=input_layer, outputs=output_layer)
```

```python
autoencoder.compile(optimizer='adam', loss='mean_squared_error')
```

```python
autoencoder.fit(input_data, input_data, epochs=10, batch_size=32)
```

```python
weights = autoencoder.get_weights()[0]  # Get the weights of the first layer (between input and hidden)

weight_mean = np.mean(weights)
weight_std = np.std(weights)

threshold = weight_mean +  1.4* weight_std

potential_predictors = []
for i in range(hidden_dim):  # Iterate over the hidden nodes
    input_nodes = np.where(weights[:, i] > threshold)[0]  # Find the input nodes with weights above the threshold
    if len(input_nodes) >= input_dim * 0.1: # Check if at least 10% of input nodes participate
        predictor = np.sum(weights[input_nodes, i] * input_data[:,input_nodes])
        potential_predictors.append(predictor)

print("Potential predictors:")
print(potential_predictors)
```

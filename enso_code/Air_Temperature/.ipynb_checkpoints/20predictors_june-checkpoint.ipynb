# Bagging Regressor:

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor

# Assuming x_train and y_train are your training data
# Adjust this based on your actual data

# Define the parameter grid
param_grid = {
    'base_estimator__max_depth': range(1, 5),
    'n_estimators': range(1, 51)
}

# Create the BaggingRegressor with DecisionTreeRegressor as the base estimator
bagging_reg = BaggingRegressor(base_estimator=DecisionTreeRegressor(), oob_score=True, random_state=0)

# Create the GridSearchCV object
grid_search = GridSearchCV(bagging_reg, param_grid, cv=3, scoring='r2')

# Fit the grid search to the data
grid_search.fit(x_train, y_train_mean)

# Get the best parameters and the corresponding score
best_params = grid_search.best_params_
best_score = grid_search.best_score_
best_max_depth = best_params['base_estimator__max_depth']
best_n_estimators = best_params['n_estimators']

print(f"Best Parameters: {best_params}")
print(f"Best Score: {best_score}")
```

```python
best_max_depth
```

```python
best_n_estimators
```

```python
def model_bag_reg(max_depth,n_estimators,x_test):
    regressor = BaggingRegressor(
          base_estimator=DecisionTreeRegressor(max_depth=max_depth),
          n_estimators=n_estimators,
          oob_score=True,
          random_state=0)
    regressor.fit(x_train,y_train_mean)
    score = regressor.score(x_train,y_train_mean)
    y_pred_br5 = regressor.predict(x_test)
    return score,y_pred_br5
```

```python
score,pred_bag_reg_mean = model_bag_reg(best_max_depth,best_n_estimators,x_test)
```

```python
pred_bag_reg_mean
```

```python
corr_bag_reg_mean = np.corrcoef(pred_bag_reg_mean, y_test_mean)
corr_bag_reg_mean
```

# Random Forest

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# Assuming x_train and y_train are your training data
# Adjust this based on your actual data

# Define the parameter grid
param_grid = {
    'n_estimators': range(1,51),  # Adjust as needed
    'max_depth': range(1,5),  # Adjust as needed
    'min_samples_split':range(1,5), # minimum number of samples required to split an internal node
    'min_samples_leaf': range(1,3)  # minimum number of samples required to be in a leaf node
}

# Create the RandomForestRegressor
random_forest_reg = RandomForestRegressor(random_state=0)

# Create the GridSearchCV object
grid_search = GridSearchCV(random_forest_reg, param_grid, cv=3, scoring='r2')

# Fit the grid search to the data
grid_search.fit(x_train, y_train_mean)

# Get the best parameters and the corresponding score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Save the best values into separate variables
best_max_depth = best_params['max_depth']
best_n_estimators = best_params['n_estimators']

print(f"Best Parameters: {best_params}")
print(f"Best Score: {best_score}")
print(f"Best max_depth: {best_max_depth}")
print(f"Best n_estimators: {best_n_estimators}")
```

```python
min_samples_leaf = best_params['min_samples_leaf']
min_samples_split = best_params['min_samples_split']
```

```python
def Randomforest(max_depth,n_estimators,min_samples_split,min_samples_leaf,x_test): 
    RF = RandomForestRegressor(
          n_estimators=n_estimators,
          max_depth=max_depth,
          min_samples_split = min_samples_split,
          min_samples_leaf = min_samples_leaf,
          random_state=0)
    RF.fit(x_train,y_train_mean)
    score = RF.score(x_train,y_train_mean)
    y_pred_rf = RF.predict(x_test)
    return score,y_pred_rf
    
```

```python
score,pred_rf_mean = Randomforest(best_max_depth,best_n_estimators,min_samples_split,min_samples_leaf,x_test)
```

```python
corr_rf_mean = np.corrcoef(pred_rf_mean, y_test_mean)
corr_rf_mean
```
